# 01 — Problem Framing

**Owner:** Person B (decision track).

**Rubric line:** KPIs + strategic alignment with the business problem.

This notebook is mostly *writing* — almost no code. Outputs a clear,
executive-readable framing that the rest of the project consistently
refers back to. Everything below should map verbatim into the slide
deck and memo at the end.


## 1.1 — Five-sentence problem brief

Cover, in five sentences:

1. **Context** — NovaBank's situation and why now.
2. **Decision to be made** — who to contact for the next retention campaign.
3. **Who cares** — operations, customer success, finance.
4. **Timeframe** — campaign cycle (e.g. monthly).
5. **Constraints** — fixed contact budget, fairness/explainability requirement.


## 1.2 — Business → analytics translation

Business problem: *"Which customers should we contact next campaign cycle?"*

Analytics problem: **Binary classification** — predict P(subscribe | features),
where subscription is the retention signal. Score every customer, rank, and
select the top K% under a contact budget.

Note explicitly why we are NOT predicting churn directly: the dataset has no
churn label; subscription is the closest available retention signal.


## 1.3 — Success measures (1–2 KPIs)

Pick two, document why:

- **Primary: PR-AUC (average precision).** ROC-AUC is misleading at ~11%
  positive rate; PR-AUC reflects what we actually care about — precision
  among the customers we contact.
- **Secondary: Expected value at top-K decile.** Translates probabilities
  into € so executives can compare against campaign budget.


## 1.4 — Draft action / policy

**Initial policy:** *"Each campaign cycle, contact the top 20% of customers by
predicted subscription probability."*

This will be refined in notebook 07. We start with 20% as the default budget;
notebook 08 will sensitivity-test 5%, 10%, 30%.


## 1.5 — What this framing buys us

- A target with real labels (no synthetic churn).
- A clean FP/FN trade-off: FP = wasted call, FN = missed engagement.
- A defensible explanation surface: "why was this customer prioritised?"
